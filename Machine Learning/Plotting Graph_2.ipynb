```python
import pandas as pd  
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('1. Regression - Module - (Housing Prices).csv')
```

```python
df.head(10)
```

```python
plt.scatter(x=df['Flat Area (in Sqft)'], y = df['Sale Price'],color = 'red')
plt.xlabel("Flat Area")
plt.ylabel("Sale Price")
plt.title("Flat Area v/s Sale Price")
plt.show()
```

```python
plt.scatter(x=df['No of Bathrooms'], y =df['Sale Price'],color = 'red')
plt.xlabel("No. of Bathrooms")
plt.ylabel("Sale price")
plt.title("No. Of Bathrooms v/s Sale Price")
plt.show()
```

```python
plt.scatter(x=df['Age of House (in Years)'], y =df['Sale Price'],color = 'red')
plt.xlabel("Age of House")
plt.ylabel("Sale price")
plt.title(" Age of House v/s Sale Price")
plt.show()
```
