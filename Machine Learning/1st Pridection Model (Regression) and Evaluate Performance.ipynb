```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
```

```python
data = pd.read_csv('Transformed_Housing_Data2.csv')
data.head()
```

```python
data['mean_sales']= data['Sale_Price'].mean()
data['mean_sales'].head()
```

```python
plt.figure(dpi=100)
k=range(0, len(data))
plt.scatter(k, data['Sale_Price'].sort_values(), color ='red', label= 'Actual Price')
plt.plot(k, data['mean_sales'].sort_values(), color = 'green', label='mean-price')
plt.xlabel('Fitted Points(Ascending)')
plt.ylabel('Sale Price')
plt.title('Overall Mean')
plt.legend()
```

```python
grades_mean= data.pivot_table(values = 'Sale_Price',columns ='Overall Grade', aggfunc =np.mean)
grades_mean
```

```python
# making new column
data['grade_mean'] = 0

# for every grade fill its mean price in new column
for i in grades_mean.columns:
     data['grade_mean'][data['Overall Grade'] == i] = grades_mean[i][0]
  
print(data['grade_mean'].head())
```

```python
gradewise_list = []
for i in range(1,11):
    k =data['Sale_Price'][data['Overall Grade']==i]
    gradewise_list.append(k)
```

```python
classwise_list = []
for i in range(1,11):
    k =data['Sale_Price'][data['Overall Grade']==i]
    classwise_list.append(k)
```

```python
plt.figure( dpi = 120, figsize = (15,9) )

#### Plotting "Sale_Price" gradewise ####
# z variable is for x-axis
z = 0
for i in range(1,11):
  # defining x-axis using z
  points = [k for k in range(z, z + len(gradewise_list[i-1]))]
  # plotting
  plt.scatter( points, # x-axis
              np.sort(gradewise_list[i-1], axis = 0) , # y-axis
              label = ('houses with overall grade', i), s = 4 )            
  z = max(points) + 1 # updating z for next plot

#### plotting Overall Mean ####
plt.scatter([q for q in range(0, z)],
         data['mean_sales'],
         color = 'red',
         label = 'Overall Mean',
         s = 8)

plt.xlabel('Fitted points (Ascending)')
plt.ylabel("Sale Price")
plt.title('Overall Mean')
plt.legend( loc = 4)
```

```python
#residual Plot 
#residual = pridection - actual
mean_difference= data['mean_sales']- data['Sale_Price']
grade_mean_difference =data['grade_mean']- data['Sale_Price']
```

```python
#Residual plot is a scatter plot of difference between prediction and Actual.
k= range(0,len(data))
l= [0 for i in range(len(data))]

plt.figure(figsize =(15,6), dpi =100)

plt.subplot(1,2,1)
plt.scatter(k, mean_difference , color= 'red', label ='Residuals' ,s=2)
plt.plot(k,l,color='green',label = 'Mean Regresion', linewidth = 3)
plt.xlabel('Fitted Points')
plt.ylabel('Residual')
plt.title('Residuals with respect to Overall Mean')

plt.subplot(1,2,2)
plt.scatter(k,grade_mean_difference, color = 'red' , label = 'Residuals' ,s =2 )
plt.plot(k,l , color='green',label='Mean_Regression',linewidth=3)
plt.xlabel('Fitted Points')
plt.ylabel('Residuals')
plt.legend()
plt.title('Residuals with respect to Overall Grade Mean')

plt.legend()
```

Therefore lets find out Mean Absolute Error (MAE)

```python
Y= data['Sale_Price']

Y_hat1=data['mean_sales']
Y_hat2=data['grade_mean']

n=len(data)
len(Y),len(Y_hat1),len(Y_hat2) ,n
```

```python
#lets calculate the error with 1st regresion i.e with Y_hat1
cost_mean = sum(abs(Y_hat1 - Y))/n
cost_mean
```

```python
#lets calculate the error with 2nd regresion i.e with Y_hat2
cost_garde_mean = sum(abs(Y_hat2 - Y))/n
cost_grade_mean
```

As we can see the second module is more correct than first 

```python
#lets try using scikit-learn lib
from sklearn.metrics import mean_absolute_error
cost_grade_mean = mean_absolute_error(Y_hat2,Y)
cost_grade_mean
```

```python
#lets find error using Mean Squared Error(MSE)
from sklearn.metrics import mean_squared_error as msa
cost_mean = msa(Y_hat1,Y)
cost_grade_mean = msa(Y_hat2,Y)
cost_mean,cost_grade_mean
```

```python
#lets find the error using Root Mean Squared Error(RMSE)
#as RMSE is not availabel in Scikit lib
from sklearn.metrics import mean_squared_error as mse
cost_mean = mse(Y_hat1,Y)**0.5
cost_grade_mean = mse(Y_hat2,Y)**0.5
cost_mean,cost_grade_mean
```
